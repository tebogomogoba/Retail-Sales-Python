```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Fetching data via Spark
retail_sales = spark.table("workspace.default.retail_sales_dataset")
display(retail_sales.limit(20))
```

```python
#transferring to pandas dataframe
retail_sales = retail_sales.toPandas()
```

```python
#changing data type of Date column from object to Date
retail_sales['Date'] = pd.to_datetime(retail_sales['Date'])
```

```python
#Creating a Year, Month and Day columns in the table
date_column = pd.DatetimeIndex(retail_sales['Date'])
retail_sales['Year'] = date_column.year
retail_sales['Month'] = date_column.month
retail_sales['Day'] = date_column.day
```

```python
display(retail_sales)
```

```python
#Sales by category 
sales_by_category = retail_sales.groupby('Product Category')['Total Amount'].sum()
display(sales_by_category)
```

```python
#PLotting a sales by category Bar Graph
plt.figure()
sales_by_category.plot(kind='bar')
plt.title('Total sales by category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
#Plotting sales trend overtime (Time index/line Graph)
sales_trend = retail_sales.groupby('Date')['Total Amount'].sum()

plt.figure()
sales_trend.plot()
plt.title('Sales Trend Overtime')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
#Sales by month using Seaborn
#Seaborn basically makes more pretty graphs

plt.figure()
sns.barplot(data=retail_sales, x= "Month", y="Total Amount")
plt.title('Sales by Month')
plt.show()
```

```python
#Sales by month using matplot
sales_by_month = retail_sales.groupby('Month')['Total Amount'].sum()
plt.figure()
sales_by_month.plot(kind="bar")
plt.title('Total sales each month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
display(sales_by_month)
```

```python
#plotting histogram using Matplot
plt.figure()
plt.hist(retail_sales['Total Amount'], bins=10)
plt.title("Distribution of Total Sales Amount")
plt.xlabel('Sales Amount')
plt.ylabel("Frequency")
plt.show()
```

```python
#Plotting a pie chart to show the share of sales for each product category using Panda

plt.figure()
sales_by_category.plot(kind='pie', autopct = '%1.1f%%')
plt.title('Sales contribution by category (Percentage share per product)')
plt.ylabel("")
plt.show()
```

```python
#plotting Sales Trend with markers using Panda and Matplotlib
plt.figure(figsize=(20,5))
plt.plot(sales_trend, marker ='o')
plt.title('Sales Trend with Markers')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
#How to Change Colours and size on the graph example usinbg Panda and Matplotlib
plt.figure(figsize=(15,5))
sales_trend.plot(color='orange')
plt.title('Sales Trend Overtime')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```
